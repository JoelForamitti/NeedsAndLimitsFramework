## Domain Fulfillment Function

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def fulfillment(grt, sat):
    return 1 - np.exp(-grt * sat)
```

```python
dfs = []
satisfaction = np.linspace(0, 2, 100)
x = 'Need satisfaction (total satisfier impact)'
y = 'Domain fulfillment'
z = 'Satiation\nrate'

for growth_rate in (1,2,3,4,5):
    df = pd.DataFrame()
    dfs.append(df)
    df[x] = satisfaction
    df[y] = fulfillment(growth_rate, satisfaction)
    df[z] = growth_rate  
    
df = pd.concat(dfs, axis=0)
```

```python
sns.set_theme()
sns.lineplot(data=df, x=x, y=y, hue=z, palette='crest')
plt.savefig(f"../Plots/domain_fulfillment.pdf", bbox_inches='tight')
```
